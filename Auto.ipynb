```python
#imports
import numpy as np
from ISLP import load_data
```

```python
#Data
auto = load_data('Auto')
auto
```

```python
#qualitative vs quantitative predictors

# List quantitative (numeric) column names
quantitative_columns = auto.select_dtypes(include=['int64', 'float64']).columns.tolist()

# List qualitative (categorical) column names
qualitative_columns = auto.select_dtypes(include=['object']).columns.tolist()

# Print the lists
print("Quantitative Columns:", quantitative_columns)
print("Qualitative Columns:", qualitative_columns)
```

```python
# Calculate the range for each quantitative predictor
ranges = {column: np.ptp(auto[column]) for column in auto.select_dtypes(include=['int64', 'float64']).columns}

# Print the ranges
for predictor, predictor_range in ranges.items():
    print(f"Range of {predictor}: {predictor_range}")
```

```python
#mean and standard deviation
statistics = {
    'Mean': auto[quantitative_columns].mean(),
    'Standard Deviation': auto[quantitative_columns].std()
}

# Print the statistics
for stat_type, values in statistics.items():
    print(f"\n{stat_type} of each quantitative predictor:")
    for predictor, value in values.items():
        print(f"{predictor}: {value}")
```

```python
# Remove the 10th through 85th observations
df_subset = auto.drop(auto.index[9:84])

# Calculate the range, mean, and standard deviation for each predictor in the subset
quantitative_columns = df_subset.select_dtypes(include=['int64', 'float64']).columns

statistics_subset = {
    'Range': np.ptp(df_subset[quantitative_columns], axis=0),
    'Mean': np.mean(df_subset[quantitative_columns], axis=0),
    'Standard Deviation': np.std(df_subset[quantitative_columns], axis=0)
}

# Print the statistics for the subset
for stat_type, values in statistics_subset.items():
    print(f"\n{stat_type} of each predictor in the subset:")
    for i, predictor in enumerate(quantitative_columns):
        print(f"{predictor}: {values[i]}")
```
